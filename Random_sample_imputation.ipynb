```python
df = pd.read_csv('../datasets/AmesHousing.csv', usecols=['Garage Yr Blt','SalePrice','Mas Vnr Area'])
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
X = df.drop(columns=['SalePrice'])
y = df['SalePrice']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train['Garage_Yr_Blt_imputed'] = X_train['Garage Yr Blt']
X_test['Garage_Yr_Blt_imputed'] = X_test['Garage Yr Blt']
```

```python
X_train['Garage_Yr_Blt_imputed'][X_train['Garage_Yr_Blt_imputed'].isnull()] = X_train['Garage Yr Blt'].dropna().sample(X_train['Garage Yr Blt'].isnull().sum()).values
# The LHS is boolean masking. The RHS, count the number of null values, sample that many values from X_train's garage yr blt col, drop the null values from the X_train's garage yr blt col. Then fill it in the imputed values column.
```

```python
X_test['Garage_Yr_Blt_imputed'][X_test['Garage_Yr_Blt_imputed'].isnull()] = X_test['Garage Yr Blt'].dropna().sample(X_test['Garage Yr Blt'].isnull().sum()).values
```

```python
X_train.sample(5)
```

```python
sns.kdeplot(X_train['Garage Yr Blt'], label='Original')
sns.kdeplot(X_train['Garage_Yr_Blt_imputed'], label='Imputed')
plt.legend()
plt.show()
```

```python
print('Original variable variance: ', X_train['Garage Yr Blt'].var())
print('Variance after random imputation: ', X_train['Garage_Yr_Blt_imputed'].var())
#Not much change, which is good
```

```python
df = pd.read_csv('../datasets/AmesHousing.csv', usecols=['Garage Type','Garage Qual','SalePrice'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
X = df
y = df['SalePrice']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train['Garage_Type_imputed'] = X_train['Garage Type']
X_test['Garage_Type_imputed'] = X_test['Garage Type']

X_train['Garage_Qual_imputed'] = X_train['Garage Qual']
X_test['Garage_Qual_imputed'] = X_test['Garage Qual']
```

```python
X_train.sample(5)
```

```python
X_train['Garage_Type_imputed'][X_train['Garage_Type_imputed'].isnull()] = X_train['Garage_Type_imputed'].dropna().sample(X_train['Garage_Type_imputed'].isnull().sum()).values
X_test['Garage_Type_imputed'][X_test['Garage_Type_imputed'].isnull()] = X_test['Garage_Type_imputed'].dropna().sample(X_test['Garage_Type_imputed'].isnull().sum()).values

X_train['Garage_Qual_imputed'][X_train['Garage_Qual_imputed'].isnull()] = X_train['Garage_Qual_imputed'].dropna().sample(X_train['Garage_Qual_imputed'].isnull().sum()).values
X_test['Garage_Qual_imputed'][X_test['Garage_Qual_imputed'].isnull()] = X_test['Garage_Qual_imputed'].dropna().sample(X_test['Garage_Qual_imputed'].isnull().sum()).values
```

```python
X_train.sample(5)
```

```python
temp = pd.concat(
    [
        X_train['Garage Type'].value_counts()/len(X_train['Garage Type'].dropna()), #Count the ratio of each garage type out of the total non null values
        X_train['Garage_Type_imputed'].value_counts()/len(X_train) #Same after imputation
    ],axis=1
)

temp.columns = ['original','imputed']
```

```python
temp
```

```python
temp = pd.concat(
    [
        X_train['Garage Qual'].value_counts()/len(X_train['Garage Qual'].dropna()),
        X_train['Garage_Qual_imputed'].value_counts()/len(X_train)
    ],axis=1
)

temp.columns = ['original','imputed']
```

```python
temp
```

```python
for category in X_train['Garage Type'].dropna().unique():
    sns.kdeplot(X_train[X_train['Garage Type']==category]['SalePrice'],label=category)
plt.legend()
plt.show()
```

```python
for category in X_train['Garage_Type_imputed'].dropna().unique():
    sns.kdeplot(X_train[X_train['Garage_Type_imputed']==category]['SalePrice'],label=category)
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,5))
for category in X_train['Garage Type'].dropna().unique():
    sns.kdeplot(X_train[X_train['Garage Type']==category]['SalePrice'],label=category, ax=axes[0])

axes[0].set_title('Original')
axes[0].legend()

for category in X_train['Garage_Type_imputed'].dropna().unique():
    sns.kdeplot(X_train[X_train['Garage_Type_imputed']==category]['SalePrice'],label=category, ax=axes[1])

axes[1].set_title('Imputed')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,5))
for category in X_train['Garage Qual'].dropna().unique():
    sns.kdeplot(X_train[X_train['Garage Qual']==category]['SalePrice'],label=category, ax=axes[0])

axes[0].set_title('Original')
axes[0].legend()

for category in X_train['Garage_Qual_imputed'].dropna().unique():
    sns.kdeplot(X_train[X_train['Garage_Qual_imputed']==category]['SalePrice'],label=category, ax=axes[1])

axes[1].set_title('Imputed')
axes[1].legend()

plt.tight_layout()
plt.show()
```
