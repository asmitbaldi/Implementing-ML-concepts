```python
df['TAILNUM'] = df['TAILNUM'].fillna('Missing')
```

```python
df['Numerical TAILNUM'] = pd.to_numeric(df['TAILNUM'],errors='coerce',downcast='integer')
```

```python
df['Numerical TAILNUM'].sample(5)
#Can't seem to find any only numeric coulmn
```

```python
df['Categorical TAILNUM'] = np.where(df['Numerical TAILNUM'].isnull(),df['TAILNUM'],np.nan)
```

```python
df['Categorical TAILNUM'].sample(5)
```

```python
#The above thing is for columns which has numerical and categorical data in one, but in different rows. eg. row1- 10, row2-A, to create different coulns for them
```

```python
df['TAILNUM_num'] = df['TAILNUM'].str.extract(r'(\d+)')
```
