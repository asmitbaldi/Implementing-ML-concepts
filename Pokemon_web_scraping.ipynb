```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36"
}
```

```python
url = 'https://pokemondb.net/pokedex/game/red-blue-yellow'
```

```python
webpage = requests.get(url,headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
soup.prettify()
```

```python
name = []
type = []
for i in soup.find_all('div', class_="infocard"):
    name.append(i.find('a',class_="ent-name").text.strip())
    type.append([x.text.strip()for x in i.find_all ('a',class_="itype")])
```

```python
name
```

```python
type
```

```python
df = pd.DataFrame({'name':name,'type':type})
```

```python
df
```

```python
new_df = df['type'].apply('  .  '.join)
```

```python
new_df
```

```python
df3 = pd.DataFrame({'name':name,'type':new_df})
```

```python
df3[['type1', 'type2']] = df3['type'].str.split(' . ', expand=True)
```

```python
df3
```

```python
def color_type(val):
    color_types = {
        'Normal': '#A8A77A',
        'Fire': '#EE8130',
        'Water': '#6390F0',
        'Electric': '#F7D02C',
        'Grass': '#7AC74C',
        'Ice': '#96D9D6',
        'Fighting': '#C22E28',
        'Poison': '#A33EA1',
        'Ground': '#E2BF65',
        'Flying': '#A98FF3',
        'Psychic': '#F95587',
        'Bug': '#A6B91A',
        'Rock': '#B6A136',
        'Ghost': '#735797',
        'Dragon': '#6F35FC'
    }

    return f'''
       background-color: {color_types.get(val, "#FFFFFF")};

        color: black;

        font-family: monospace;

        font-weight: medium;

        border-radius: 0px;
    '''
```

```python
df4 = pd.DataFrame({'name':name,'type1':df3['type1'],'type2':df3['type2']})
```

```python
df4['type1'] = df4['type1'].str.strip()
df4['type2'] = df4['type2'].str.strip()
df4 = df4.fillna('   ')
```

```python
df5 = df4.style.hide(axis='index').map(color_type, subset=['type1', 'type2']).set_table_styles(
[

{
    'selector': 'th,td',
    'props': [
        ('border', '1px solid black'),
        ('padding', '7px'),
        ('font-family', 'monospace')
    ]
},
{
    'selector': 'table',
    'props': [
        ('border-collapse', 'collapse'),
        ('margin-left', 'auto'),
        ('margin-right', 'auto')
    ]
},



]



)
```

```python
df5 = df5.set_properties(

    subset=['name'],

    **{'font-family': 'monospace'}

)
```

```python
df5.to_html("pokemon.html", index = False)
```
