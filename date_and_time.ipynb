```python
date = date[['signup_date','region','country']]
```

```python
date.head()
```

```python
date.info()
```

```python
date['signup_date'] = pd.to_datetime(date['signup_date'])
```

```python
date.info()
```

```python
date['year'] = date['signup_date'].dt.year
date['month'] = date['signup_date'].dt.month
date['day'] = date['signup_date'].dt.day
date['month_name'] = date['signup_date'].dt.month_name()
date['day_name'] = date['signup_date'].dt.day_name()
date['day_of_week'] = date['signup_date'].dt.day_of_week
```

```python
date
```

```python
date['day_is_a_weekend'] = np.where(date['day_name'].isin(['Saturday','Sunday']),1,0)
```

```python
date.sample(5)
```

```python
date['day_is_a_weekday'] = np.where(date['day_name'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday']),1,0)
```

```python
date.sample(5)
```

```python
date['quarter'] = date['signup_date'].dt.quarter
```

```python
date['quarter'].info()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
```

```python
date
```
